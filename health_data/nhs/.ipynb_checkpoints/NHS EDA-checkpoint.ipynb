```python
# Load in the datasets (concatenated in the nhs_load_data.ipynb notebookn )
mortality_nhs = pd.read_csv('mortality_hospital/concat_final/hospital_mortality_rate.csv', encoding='utf-8')
admission_nhs = pd.read_csv('hospital_sets/csv/concat/nhs_hospital_admissions_2008_2020.CSV', encoding='utf-8')
```

```python
# function I had to create to treat the float values (likely from empty rows)
def bespoke_map_column_type(col):
    try:
        return float(col.strip().replace(',', '').replace('-', '0').replace('*', '0'))
    
    # capture cases where value is not a float but is also not a string
    except ValueError:
        return np.nan
    
    # capture case where pandas cannot perform string operations on the blank spaces
    except AttributeError:
        return np.nan
```

```python
# conver spells and expected to float for now... 
mortality_nhs['SPELLS'] = mortality_nhs['SPELLS'].str.replace('*', '0').astype('float')
mortality_nhs['OBSERVED'] = mortality_nhs['OBSERVED'].str.replace('*', '0').astype('float')
mortality_nhs['EXPECTED'] = mortality_nhs['EXPECTED'].str.replace('*', '0').astype('float')

admission_nhs['Diseases of the respiratory system'] = admission_nhs['Diseases of the respiratory system'].map(bespoke_map_column_type)
admission_nhs['Diseases of the circulatory system'] = admission_nhs['Diseases of the circulatory system'].map(bespoke_map_column_type)

admission_nhs['Commissioning Region/Provider'].fillna('Not Specified', inplace=True)
```

## Mortality Dataset Analysis

```python
# do some groupby and  plots, also reference which health indicators we are interested in
master_dict = {
    73: 'Pneumonia (excluding TB/STD)',
    74: 'Acute bronchitis',
    75: 'COPD & bronchiectasis',
    76: 'Asthma',
    77: 'Aspiration pneumonitis; food/vomitus',
    78: 'Pleurisy; pneumothorax; pulmonary collapse',
    79: 'Respiratory failure; insufficiency; arrest (adult)',
    80: 'Lung disease due to external agents',
    81: 'Cystic fibrosis, Other lower respiratory disease',
    82: 'Upper respiratory disease, Diseases of mouth (non dental)',
    96: 'Heart valve disorder',
    97: 'Peri-; endo-; and myocarditis; cardiomyopathy ',
    98: 'Essential hypertension, Hypertension with complications and secondary hypertension',
    99: 'Essential hypertension, Hypertension with complications and secondary hypertension',
    100: 'Acute myocardial infarction',
    101: 'Coronary atherosclerosis and other heart disease',
    102: 'Nonspecific chest pain',
    103: 'Pulmonary heart disease',
    104: 'Other and ill-defined heart disease',
    105: 'Conduction disorders',
    106: 'Cardiac dysrhythmias',
    107: 'Cardiac arrest and ventricular fibrillation',
    108: 'Congestive heart failure; nonhypertensive',
    109: 'Acute cerebrovascular disease',
    110: 'Occlusion or stenosis of precerebral arteries, Other and ill-defined cerebrovascular disease',
    111: 'Occlusion or stenosis of precerebral arteries, Other and ill-defined cerebrovascular disease',
    112: 'Occlusion or stenosis of precerebral arteries, Other and ill-defined cerebrovascular disease',
    114: 'Peripheral and visceral atherosclerosis',
    115: 'Aortic; peripheral; and visceral artery aneurysms',
    116: 'Aortic and peripheral arterial embolism or thrombosis',
    117: 'Other circulatory disease',
}



def plot_indicator_df(df):
    # Important graph is to plot the different indicators values by year (so use a sum instead of count)
    grouped_sum = df.groupby(['Date Period', 'DIAGNOSIS GROUP']).sum()['OBSERVED'].reset_index()
    
    # reduce the quarterly/yearly observations to one for now (so we can make sense of the plots)
    grouped_sum = grouped_sum[grouped_sum['Date Period'].str.contains('Jan')]
    grouped_sum['DIAGNOSIS GROUP'] = grouped_sum['DIAGNOSIS GROUP'].map(lambda x: master_dict[x] if x in master_dict.keys() else None)

    # we will use seaborn here (for pandas will need another pivot table, where we break out indicator name)
    # set the figure size with rc argument for seaborn (stack overflow)
    sns.set(rc={'figure.figsize':(11,8)})

    sns.lineplot(data=grouped_sum, x='Date Period', y='OBSERVED', hue='DIAGNOSIS GROUP', legend='full')

    # Put the legend out of the figure using matlibplot (stackoverflow...)
    plt.legend((bbox_to_anchor)=(1.03,1), loc=2, borderaxespad=1)
    plt.show()
    
def plot_indicator_df_admission(df):
    # Important graph is to plot the different indicators values by year (so use a sum instead of count)
    grouped_sum = df.groupby(['Date Period', 'DIAGNOSIS GROUP']).sum()['OBSERVED'].reset_index()

    # we will use seaborn here (for pandas will need another pivot table, where we break out indicator name)
    # set the figure size with rc argument for seaborn (stack overflow)
    sns.set(rc={'figure.figsize':(11,8)})

    sns.lineplot(data=grouped_sum, x='Date Period', y='OBSERVED', hue='DIAGNOSIS GROUP', legend='full')

    # Put the legend out of the figure using matlibplot (stackoverflow...)
    plt.legend((bbox_to_anchor)=(1.03,1), loc=2, borderaxespad=1)
    plt.show()

# very hacky, combines multiple instances of trust names to a standardised name format
def bespoke_method_2(arggg):
    if 'hillingdon' in arggg.strip().lower():
        return 'Hillingdon Trust'
    elif 'ealing' in arggg.strip().lower():
        return 'Ealing NHS Trust'
    elif 'ashford' in arggg.strip().lower():
        return 'Ashford NHS Trust'
    elif 'west middlesex' in arggg.strip().lower():
        return 'West Middlesex NHS Trust'
    elif 'frimley' in arggg.strip().lower():
        return 'Frimley NHS Trust'
    elif 'berkshire' in arggg.strip().lower():
        return 'Berkshire NHS Trust'
    elif 'hounslow' in arggg.strip().lower():
        return 'Hounslowe NHS Trust'
    elif 'teddington' in arggg.strip().lower():
        return 'Teddington NHS Trust'
    elif 'richmond' in arggg.strip().lower():
        return 'Richmond NHS Trust'
    elif 'clayponds' in arggg.strip().lower():
        return 'Clayponds NHS Trust'
    elif 'cygnet' in arggg.strip().lower():
        return 'Cygnet NHS Trust'
    elif 'cassel' in arggg.strip().lower():
        return 'Cassel NHS Trust'
    elif 'london north west university healthcare' in arggg.strip().lower():
        return 'London North West University Healthcare' 
    elif 'north west london hospitals nhs trust' in arggg.strip().lower():
        return 'North West London Hospitals NHS Trust'
    else:
        return 'Trust Not Specified'
```

## Admission Dataset Analysis

```python
# hack for now - get the year column from the data (originally an int which we made in the last notebook)
date_dict = {809:'2008', 
             910:'2009', 
             1011:'2010', 
             1112:'2011', 
             1213:'2012', 
             1314:'2013', 
             1415:'2014', 
             1516:'2015', 
             1617:'2016', 
             1718:'2017', 
             1819:'2018', 
             1920:'2019', 
            }

# groupby date and sum to get the admissions per year, then plot 
grouped_sum = admission_nhs.groupby(['Date Period']).sum()['Diseases of the respiratory system'].reset_index()
grouped_sum['Date Period'] = grouped_sum['Date Period'].map(lambda x: date_dict[x])


# plot
sns.lineplot(data=grouped_sum, x='Date Period', y='Diseases of the respiratory system', legend='full')

plt.show()
```

```python
# groupby date and sum to get the admissions per year, then plot 
grouped_sum1 = admission_nhs.groupby(['Date Period']).sum()['Diseases of the circulatory system'].reset_index()
grouped_sum1['Date Period'] = grouped_sum1['Date Period'].map(lambda x: date_dict[x])

# plot
sns.lineplot(data=grouped_sum1, x='Date Period', y='Diseases of the circulatory system', legend='full')


plt.show()
```

```python
# we also want to see a view of admissions for hospital trusts we are interested in.
# query the admissions df by areas around heathrow to return hospital types close to the airport

# I got areas by googling hospitals near heathrow, then searching for trusts in areas using str.contains below
hospitals_1 = ['ashford', 'hillingdon', 'ealing', 'west middlesex', 'frimley', 'berkshire', 'hounslow', 'teddington', 'richmond', 'clayponds', 'slough', 'cygnet', 'cassel', 'london north west university healthcare', 'north west london hospitals nhs trust']
admission_nhs['Commissioning Region/Provider'].fillna('NOT SPECIFIED - OMIT OR OTHERWISE', inplace=True)

# restrict df to trusts in our hospital list above
tt = admission_nhs[admission_nhs['Commissioning Region/Provider'].str.lower().str.contains('|'.join(hospitals_1))]
tt['Commissioning Region/Provider'] = tt['Commissioning Region/Provider'].apply(bespoke_method_2)

# get group by hospital to check whether hospital trust have data consistency 
grouped_sum2 = tt.groupby(['Date Period', 'Commissioning Region/Provider']).sum()['Diseases of the respiratory system'].reset_index()
grouped_sum2['Date Period'] = grouped_sum2['Date Period'].map(lambda x: date_dict[x])


# sns.lineplot(data=grouped_sum, x='Date Period', y='Diseases of the respiratory system', legend='full')
sns.lineplot(data=grouped_sum2, x='Date Period', y='Diseases of the respiratory system', hue='Commissioning Region/Provider', legend='full')

# Put the legend out of the figure using matlibplot (stackoverflow...)
plt.legend((bbox_to_anchor)=(1.03,1), loc=2, borderaxespad=1)

plt.show()
```

```python
# even closer view - just ealing and hillingdon hospital trusts
hospitals_2 = ['ealing', 'hillingdon']     

tt = admission_nhs[admission_nhs['Commissioning Region/Provider'].str.lower().str.contains('|'.join(hospitals_2))]  
tt['Commissioning Region/Provider'] = tt['Commissioning Region/Provider'].apply(bespoke_method_2)

# get group by hospital to see another view
grouped_sum2 = tt.groupby(['Date Period', 'Commissioning Region/Provider']).sum()['Diseases of the respiratory system'].reset_index()
grouped_sum2['Date Period'] = grouped_sum2['Date Period'].map(lambda x: date_dict[x])
grouped_sum2['Date Period'] = pd.to_datetime(grouped_sum2['Date Period'], format='%Y')

# plot 
sns.lineplot(data=grouped_sum2, x='Date Period', y='Diseases of the respiratory system', hue='Commissioning Region/Provider', legend='full')
plt.legend((bbox_to_anchor)=(1.03,1), loc=2, borderaxespad=1)    # Put the legend out of the figure using matlibplot (stackoverflow...)
plt.show()
```
